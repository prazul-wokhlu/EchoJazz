```python
from music21 import *
import os
import glob
```

```python
def get_chords(filename):

    s = converter.parse(filename)
    m = s.parts[0].getElementsByClass(stream.Measure)

    i = 0
    chordChart = ''

    while i < len(m):
        c = m[i].getElementsByClass(harmony.ChordSymbol)

        if(len(c) == 0):
            chordChart += '{:1} '.format('')
        if(len(c) == 1):
            c0=c[0].figure
            chordChart += '{:1} '.format(c0)
        if(len(c) == 2):
            c0=c[0].figure
            c1=c[1].figure
            chordChart += '{:1} '.format(c0, c1)
        if(len(c) == 3):
            c0=c[0].figure
            c1=c[1].figure
            c2=c[2].figure
            chordChart += '{:1} '.format(c0, c1, c2)
        if(len(c) == 4):
            c0=c[0].figure
            c1=c[1].figure
            c2=c[2].figure
            c3=c[3].figure
            chordChart += '{:1} '.format(c0, c1, c2, c3)
        i += 1
    
    chordChart = chordChart.replace("-","b")
    chordChart = chordChart.replace(' alter ', 'aug')
    chordChart = chordChart.replace(' add ','')
    
    return chordChart[:len(chordChart)-1]
    
  
```

```python
import re, string

punc_regex = re.compile('[{}]'.format(re.escape(string.punctuation)))

def strip_punc(corpus):
    return punc_regex.sub('', corpus)


```
